# Presentation Points to make

## 1. Who Is Making This Decision? What Action Will They Take?

**Decision-maker:** The **NY State Department of Health** (or NY State of Health, the state's official insurance marketplace), which controls the annual Medicaid and ACA outreach budget and funds county-level navigator programs.

**The action:** They allocate a **$1M outreach budget** across up to 10 counties by contracting navigator organizations, community health workers, and enrollment events. The model tells them exactly which 10 counties to fund and at what level — in this solution, $100k each.

**Talking point:** Frame this as a real operational decision that happens every year during ACA open enrollment season. The DOH has a finite budget and dozens of counties to consider — this model provides a principled, data-driven allocation rather than a political or ad-hoc one.

## 2. Translate the Objective Into Plain Language

**In math:** $\max \sum_i \alpha_i x_i$

**In words:** "We prioritize counties where each outreach dollar converts the most uninsured people into enrolled individuals."

$\alpha_i = k \cdot \text{PCTUI}_i$ measures the **return on investment per $1k spent**: counties with higher uninsured rates have a larger pool of people facing enrollment barriers (informational, administrative, language), so each navigator dollar goes further there.

The model is essentially ranking counties by outreach ROI and selecting the top K, subject to budget and capacity limits.

**Talking point:** This is not about helping the "worst-off" county in isolation — it is about maximizing total newly insured statewide. The equity constraint is how we separately guarantee that underserved counties are not ignored even if they are small.

## 3. Equity Constraint: With vs. Without

**The counterintuitive result:** The equity constraint ($\sum u_i y_i \geq R = 3$) is **non-binding** in our optimal solution. All 10 selected counties already have uninsured rates above the statewide median — they are all underserved by definition.

**Directly answering the three sub-questions:**

| Question | Answer |
|---|---|
| What happens without the constraint? | The same 10 counties are selected — no change. |
| Does it reduce efficiency? | No. Zero efficiency cost in the current solution. |
| Does it change which counties are selected? | No. |

**Why this happens:** The objective function $\max \sum \alpha_i x_i$ naturally gravitates toward high-PCTUI counties, and high-PCTUI counties are by definition underserved. Economic efficiency and equity align here.

**Why the constraint still matters — the talking point:** The equity constraint is a safeguard for edge cases. If the budget were very small, a handful of large urban counties might dominate the solution even with moderate uninsured rates (due to scale). The equity constraint guarantees rural and underserved representation regardless. In the current parameterization the market does the right thing on its own, but the constraint ensures that holds even if parameters shift.

### Magnitude of Impact

- **Total newly insured:** **2,788 individuals**
- **All 10 selected counties are underserved** (above median uninsured rate)

### The Recommendation

> Concentrate the outreach budget in high-uninsured-rate counties that are large enough to absorb meaningful program investment. Do not chase the highest uninsured rate alone — population scale determines whether that rate translates to real impact.

### The Yates County Hook (memorable for Q&A)

Yates County has the **single highest uninsured rate in New York State (17.8%, $\alpha = 3.56$)**, yet it is not selected. The enrollment capacity constraint exposes why.

Yates is a small rural county (NUI ≈ 579). Its enrollment capacity is:

$$c_{\text{Yates}} = \frac{0.20 \times 579}{3.56} \approx \$32.5\text{k}$$

It can only absorb ~$32.5k in outreach before hitting the operational ceiling — not the full $100k slot. Its maximum contribution is therefore:

$$\alpha_{\text{Yates}} \times c_{\text{Yates}} = 3.56 \times 32.5 \approx \textbf{116 newly insured}$$

Sullivan County, by contrast, has a lower uninsured rate (12.1%, $\alpha = 2.42$) but a large enough population to absorb the full $100k allocation:

$$\alpha_{\text{Sullivan}} \times 100 = 2.42 \times 100 = \textbf{242 newly insured}$$

Sullivan delivers **more than twice the enrollments** of Yates, so the solver prefers it.

**The insight for the audience:** A high uninsured rate is necessary but not sufficient for selection. A county must also have the population scale to absorb a viable program investment. This is a direct analog to real-world program design — small rural counties often need different intervention models (e.g., mobile units, regional partnerships) rather than the same navigator model applied to urban counties.
